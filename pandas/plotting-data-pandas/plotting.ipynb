```python
hw = pd.read_csv("./hw.csv",index_col=0)
print(hw.columns.tolist())
hw.columns = hw.columns.str.strip().str.replace('"', '')
print(hw.columns.tolist())
print(hw.head())
print(hw.columns)
hw["Height(cm)"] = hw["Height(Inches)"] * 2.54
print(hw.head())

hw.drop(columns=['Height(Inches)']).plot()
print(hw.columns)

plt.show()


plt.scatter(hw["Height(cm)"], hw["Weight(Pounds)"], alpha=0.7)
plt.title("Height vs Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (Pounds)")
plt.grid(True)
plt.show()
```
